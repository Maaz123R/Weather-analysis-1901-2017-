```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("/content/weather.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
df.duplicated().sum()
```

```python
months = ['JAN','FEB','MAR','APR','MAY','JUN','JUL','AUG','SEP','OCT','NOV','DEC']

df['YEAR_AVG_TEMP'] = df[months].mean(axis=1)

df[['YEAR', 'YEAR_AVG_TEMP']].head()
```

```python
plt.plot(df['YEAR'], df['YEAR_AVG_TEMP'])
plt.xlabel('YEAR')
plt.ylabel('TEMPERATURE')
plt.title('Average Annual Temperature Trend (19012017)')
plt.tight_layout()
plt.savefig("graphs/Average Annual Temperature Trend (19012017)")
plt.show()
```

```python
monthly_avg = df[months].mean()
monthly_avg
monthly_avg.plot(kind='bar')
plt.plot(months,monthly_avg)
plt.xlabel('MONTH')
plt.ylabel('TEMPERATURE')
plt.title('Average Monthly Temperature')
plt.savefig("graphs/Average Monthly Temperature")
plt.tight_layout()
plt.show()
```

```python
JAN_AVG = df['JAN'].mean()
FEB_AVG = df['FEB'].mean()
MAR_AVG = df['MAR'].mean()
APR_AVG = df['APR'].mean()
plt.bar(['JAN','FEB','MAR','APR'],[JAN_AVG,FEB_AVG,MAR_AVG,APR_AVG])
plt.xlabel('MONTH')
plt.ylabel('TEMPERATURE')
plt.title('Average 4 Months Temperature')
plt.savefig("graphs/Average 4 Months Temperature")
plt.tight_layout()
plt.show()
```
